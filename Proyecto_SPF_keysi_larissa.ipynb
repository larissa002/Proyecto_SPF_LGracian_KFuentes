# Simulación y Análisis del Impacto Financiero de Colisiones Severas y Fatales

## Objetivos.

### 1.1 Objetivo general.
- Desarrollar un modelo de simulación financiera para analizar el impacto económico de colisiones severas y fatales en distintas regiones geográficas.

### 1.2 Objetivos específicos
1. Recolectar y procesar datos históricos de colisiones severas y fatales, así como de variables demográficas y de tráfico.
2. Desarrollar algoritmos y modelos matemáticos para simular el comportamiento de las variables relacionadas con las colisiones, como la tasa de colisiones por 100,000 personas.
3. Implementar técnicas de validación cruzada y análisis de errores para validar el modelo de simulación desarrollado.
4. Realizar simulaciones computacionales para evaluar diferentes escenarios de colisiones severas y fatales en distintas regiones geográficas, considerando diferentes modos de transporte como vehículos, motocicletas, peatones y ciclistas.
5. Analizar los resultados obtenidos de las simulaciones para identificar patrones, tendencias y factores de riesgo asociados a las colisiones severas y fatales.
6. Proponer recomendaciones basadas en los resultados del análisis para mejorar la seguridad vial y reducir el impacto económico de las colisiones en las distintas regiones geográficas.

### 1. 3 Definición del problema

El escenario de negocio que se va a modelar es el impacto económico de colisiones severas y fatales en distintas regiones geográficas. Se busca entender cómo estas colisiones afectan tanto a la población como a la infraestructura vial, con el fin de tomar medidas preventivas y de gestión de riesgos.

**Supuestos:**
1. Las colisiones severas y fatales tienen un impacto económico significativo en términos de costos médicos, pérdida de productividad y daños a la propiedad.
2. Existe una relación entre la densidad de población, el volumen de tráfico y la tasa de colisiones en una determinada región geográfica.
3. Los datos históricos de colisiones y variables demográficas son representativos de la realidad actual y pueden utilizarse para predecir futuros escenarios.

**Nodos o Escenarios a Simular:**
1. **Recolección de Datos:** Obtener datos históricos de colisiones severas y fatales, así como de variables demográficas y de tráfico.
2. **Procesamiento de Datos:** Limpiar, procesar y preparar los datos para su análisis.
3. **Desarrollo del Modelo:** Diseñar y desarrollar un modelo de simulación financiera que integre los datos de colisiones con variables demográficas y de tráfico.
4. **Validación del Modelo:** Utilizar técnicas de validación cruzada y análisis de errores para verificar la precisión y fiabilidad del modelo.
5. **Simulación Computacional:** Realizar simulaciones para evaluar el impacto económico de las colisiones severas y fatales en diferentes regiones geográficas y escenarios de tráfico.
6. **Análisis de Resultados:** Analizar los resultados de las simulaciones para identificar patrones, tendencias y factores de riesgo asociados a las colisiones.
7. **Propuestas y Recomendaciones:** Proponer recomendaciones basadas en los resultados del análisis para mejorar la seguridad vial y reducir el impacto económico de las colisiones en las distintas regiones geográficas.

### 1.4 Nodos y variables que se decidieron simular y porqué

1. **Colisiones severas y fatales por tipo de transporte**: Este nodo implicaría modelar la ocurrencia de colisiones severas y fatales para cada uno de los seis tipos de transporte (carro/pickup, camión, autobús, motocicleta, peatón y ciclista). Cada tipo de transporte podría tener su propio modelo de simulación para estimar la cantidad de colisiones y su impacto financiero.

2. **Tasa de colisiones por ubicación geográfica**: Este nodo involucraría modelar la tasa de colisiones (severas o fatales) por cada 100,000 personas en diferentes geografías, como ciudades, condados, regiones metropolitanas, etc. Esto podría ayudar a identificar áreas de alto riesgo que podrían requerir intervenciones específicas.

3. **Comparación entre diferentes áreas geográficas**: Este nodo implicaría comparar las tasas de colisión entre diferentes áreas geográficas, como ciudades, condados o regiones metropolitanas. Esto podría ayudar a identificar patrones de comportamiento y diferencias regionales que podrían influir en la estrategia de inversión.

4. **Variación temporal de las tasas de colisión**: Este nodo implicaría modelar la variación temporal de las tasas de colisión a lo largo de los años disponibles (2002-2010). Esto podría ayudar a identificar tendencias a lo largo del tiempo y prever posibles cambios en el futuro.

Estos nodos representan diferentes aspectos de la ocurrencia de colisiones severas y fatales, así como de las tasas de colisión en diferentes contextos geográficos y temporales. Modelar estos nodos en un proyecto de simulación financiera permitiría explorar el impacto económico de las colisiones y desarrollar estrategias de inversión y políticas de seguridad vial más efectivas.

### 1.5 Definición de hipótesis y supuestos.

Definición de hipótesis y supuestos:

**Hipótesis:**
1. **Hipótesis 1:** La tasa de colisiones severas y fatales se ve influenciada por factores como la densidad de población, el volumen de tráfico y las características de la infraestructura vial.
2. **Hipótesis 2:** Existe una relación directa entre el número de colisiones y el índice de desarrollo económico de una región.
3. **Hipótesis 3:** El comportamiento del tráfico y la frecuencia de colisiones siguen patrones estacionales y diarios predecibles.
4. **Hipótesis 4:** Las políticas de seguridad vial implementadas por las autoridades tienen un impacto significativo en la reducción de la tasa de colisiones severas y fatales.
5. **Hipótesis 5:** La edad, el sexo y la experiencia del conductor son variables que influyen en la probabilidad de estar involucrado en una colisión.

**Supuestos:**
1. **Supuesto 1:** Los datos históricos de colisiones y variables demográficas son precisos y representativos de la realidad.
2. **Supuesto 2:** La infraestructura vial y las condiciones de tráfico permanecen constantes a lo largo del período de estudio.
3. **Supuesto 3:** La población y el comportamiento del tráfico en el futuro seguirán patrones similares a los observados en el pasado.
4. **Supuesto 4:** La metodología utilizada para la recolección y análisis de datos es válida y confiable.
5. **Supuesto 5:** Las políticas y regulaciones de seguridad vial se mantienen estables durante el período de estudio y no se producen cambios significativos en la legislación.
