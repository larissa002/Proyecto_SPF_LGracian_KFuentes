```python
# Importación de datos y librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
from collections import Counter
import warnings
warnings.filterwarnings("ignore")
import statsmodels.api as sm
from sklearn.neighbors import KernelDensity
from scipy.stats import gaussian_kde
```

```python
# Lectura de datos desde el archivo 'road_traffic.xlsx'
data_raw = pd.read_excel("road_traffic.xlsx")
dt = pd.DataFrame(data_raw)
```

```python
# Limpieza de datos para obtener el nodo 1 
# Filtrar las colisiones fatales
fatal_collisions = dt[dt['severity'] == 'Killed']
severa_collisions = dt[dt['severity'] == "Severe Injury"]

# Agrupar los datos por tipo de transporte y contar la cantidad de colisiones fatales
fatal_collisions_by_transport = fatal_collisions.groupby('mode').size()
severa_collisions_by_transport = severa_collisions.groupby('mode').size()

# Imprimir los resultados
print("Colisiones fatales por tipo de transporte:")
print(fatal_collisions_by_transport)

# Imprimir los resultados
print("Colisiones severas por tipo de transporte:")
print(severa_collisions_by_transport)
```

```python
# Crear histograma de probabilidad de colisiones fatales por tipo de transporte
plt.figure(figsize=(10, 6))
sns.histplot(fatal_collisions_by_transport, kde=True, stat='probability', bins=10)
plt.title('Histograma de Probabilidad de Colisiones Fatales por Tipo de Transporte')
plt.xlabel('Colisiones Fatales')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.show()
```

```python

# Definir datos de colisiones fatales por tipo de transporte
fatal_collisions_data = fatal_collisions_by_transport.values

# Definir límites del dominio
Domain = (min(fatal_collisions_data) - 1, max(fatal_collisions_data) + 1)

# Función de densidad de probabilidad para el método de aceptación-rechazo (opcional)
def f(x):
    # Aquí puedes definir tu propia función de densidad de probabilidad si es necesario
    return kde.evaluate(x)

# Función de aceptación-rechazo (aceptar si U < f(x) / M*g(x))
def acep_rechazo(N, Domain, f, M, g):
    samples = []
    count = 0
    while count < N:
        x = np.random.uniform(low=Domain[0], high=Domain[1])
        u = np.random.uniform(low=0, high=1)
        if u < f(x) / (M * g(x)):
            samples.append(x)
            count += 1
    return samples

# Definir grid para KDE
x_grid = np.linspace(min(fatal_collisions_data) - 1, max(fatal_collisions_data) + 1, 1000)

# Calcular KDE
kde = gaussian_kde(dataset=fatal_collisions_data, bw_method=0.139)
pdf = kde.evaluate(x_grid)

# Número de datos aleatorios a generar
N = 10000

# Generar datos aleatorios utilizando aceptación-rechazo
x_random = acep_rechazo(N, Domain, kde.evaluate, max(pdf), kde.evaluate)

# Graficar KDE y datos generados con aceptación-rechazo
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x_grid, pdf, color='orange', alpha=0.7, lw=3, label='KDE')
ax.hist(x_random, bins=60, density=True, alpha=0.5, label='Generados con Acep_Rechazo')
ax.set_xlim(min(fatal_collisions_data) - 1, max(fatal_collisions_data) + 1)

plt.title('Kernel Density e Histograma AyR')
plt.legend()
plt.show()
```

```python
# Parámetros de la distribución Beta ajustada
params_beta = st.beta.fit(fatal_collisions_data)

# Número de simulaciones de Monte Carlo
n_simulations = 1000

# Almacenar las varianzas calculadas en cada simulación
variances = []

# Generar muestras aleatorias y calcular la varianza en cada simulación
for _ in range(n_simulations):
    sampled_data = st.beta.rvs(*params_beta, size=len(fatal_collisions_data))
    variances.append(np.var(sampled_data))

# Calcular la media de las varianzas obtenidas en las simulaciones
monte_carlo_variance = np.mean(variances)

print("Varianza mediante Monte Carlo:", monte_carlo_variance)
```

```python

def importance_sampling_monte_carlo(f, p, num_samples):
    # Generar muestras aleatorias de la distribución de importancia p(x)
    samples = p.rvs(size=num_samples)
    
    # Calcular el valor de h(x) para cada muestra
    h_values = f(samples) / p.pdf(samples)
    
    # Estimar la media utilizando el promedio de h(x)
    estimated_mean = np.mean(h_values)
    
    return estimated_mean

# Definir la función de densidad de probabilidad f(x) que se quiere estimar (por ejemplo, una distribución gamma)
f = st.gamma(a=2)  # Por ejemplo, una distribución gamma con parámetro de forma a=2

# Definir la distribución de importancia p(x) (por ejemplo, una distribución normal)
p = st.norm(loc=2, scale=1)  # Por ejemplo, una distribución normal con media 2 y desviación estándar 1

# Número de muestras para generar
num_samples = 10000

# Estimar la media utilizando el método de Monte Carlo con muestreo por importancia
estimated_mean = importance_sampling_monte_carlo(f.pdf, p, num_samples)

print("Media estimada:", estimated_mean)
```

```python
# Definir el nivel de significancia alpha
alpha = 0.05  # Por ejemplo, un nivel de significancia del 5%

# Definir el valor crítico basado en el nivel de significancia y la distribución de importancia
valor_critico = p.ppf(1 - alpha)  # Percentil correspondiente al nivel de significancia

# Comparar el valor crítico con la media estimada
if estimated_mean > valor_critico:
    print("Se rechaza la hipótesis nula (H0)")
    print("La media estimada es significativamente mayor que el valor crítico.")
else:
    print("No se rechaza la hipótesis nula (H0)")
    
```

```python
def red_var_montecarlo(data, N):
    # Calcula la media muestral
    sample_mean = np.mean(data)
    
    # Genera N muestras aleatorias
    samples = np.random.choice(data, size=N, replace=True)
    
    # Calcula la media muestral de las muestras generadas
    sample_mean_mc = np.mean(samples)
    
    # Calcula el error relativo
    relative_error = np.abs((sample_mean - sample_mean_mc) / sample_mean) * 100
    
    return sample_mean, sample_mean_mc, relative_error

# Llamada a la función para calcular la reducción de varianza
media_muestral, media_muestral_mc, error_relativo = red_var_montecarlo(fatal_collisions_by_region.values, 10000)

print("La media muestral estimada es:", media_muestral)
print("La media muestral estimada con reducción de varianza es:", media_muestral_mc)
print("Error relativo:", error_relativo)
```

### NODO 4: Variación temporal de las tasas de colisión

```python
# Leer los datos desde el archivo 'road_traffic.xlsx'
data_raw = pd.read_excel("road_traffic.xlsx")
dt = pd.DataFrame(data_raw)

# Eliminar filas con valores vacíos
dt.dropna(inplace=True)

# Convertir la columna 'reportyear' a formato de fecha
dt['reportyear'] = pd.to_datetime(dt['reportyear'], format='%Y')

# Agrupar los datos por año y calcular la tasa de colisión promedio de cada año
collision_rates_by_year = dt.groupby(dt['reportyear'].dt.year)['avmtrate'].mean()

# Graficar la variación temporal de las tasas de colisión a lo largo de los años
plt.figure(figsize=(10, 6))
plt.plot(collision_rates_by_year.index, collision_rates_by_year.values, marker='o', linestyle='-')
plt.title('Variación Temporal de las Tasas de Colisión (2002-2010)')
plt.xlabel('Año')
plt.ylabel('Tasa de Colisión')
plt.grid(True)
plt.xticks(collision_rates_by_year.index)
plt.show()
```
